```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
barWidth = 0.15
fig = plt.subplots(figsize=(18, 12))
ttl = plt.title("Sorted Input", y=1.02, fontsize=20, weight="bold")

input_10000 = [1030005, 100151, 127992]
input_30000 = [1090005, 360186, 383992]
input_50000 = [1150005, 600186, 639992]
input_100000 = [1300005, 1200186, 1279992]
input_300000 = [1900005, 4200221, 3839992]
input_500000 = [2500005, 7000221, 6399992]

br1 = np.arange(len(input_10000))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
br6 = [x + barWidth for x in br5]

plt.bar(br1, input_10000, width=barWidth, label='input_10000')
plt.bar(br2, input_30000, width=barWidth, label='input_30000')
plt.bar(br3, input_50000, width=barWidth, label='input_50000')
plt.bar(br4, input_100000, width=barWidth, label='input_100000')
plt.bar(br5, input_300000, width=barWidth, label='input_300000')
plt.bar(br6, input_500000, width=barWidth, label='input_500000')

plt.xticks([r + 2*barWidth for r in range(len(input_10000))], ['Counting Sort', 'Radix Sort', 'Flash Sort'], fontsize=13)

legend_labels = ['Size: 10000', 'Size: 30000', 'Size: 50000', 'Size: 100000', 'Size: 300000', 'Size: 500000']
legend_markers = [plt.Line2D([0], [0], marker='s', color='w', label=label, markersize=10, markerfacecolor='C{}'.format(idx)) for idx, label in enumerate(legend_labels)]

legend = plt.legend(handles=legend_markers, loc="lower center", ncol=6, bbox_to_anchor=(0.5, -0.1), fontsize=15)

plt.grid(axis='y')
plt.show()
```
